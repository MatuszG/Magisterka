```python
import torch, os
from transformers import PreTrainedTokenizerFast, GPT2LMHeadModel, GPT2TokenizerFast, GPT2Tokenizer, GPTNeoForCausalLM, AutoTokenizer
path = os.path.join("Magisterka", "gpt2-large")
# path = os.path.join("gpt2-large")

def load_model(model_path):
    model = GPT2LMHeadModel.from_pretrained(model_path, torch_dtype=torch.float16).cuda()
    # model = model.bfloat16().cuda()
    # model.half()
    # model.to('cuda')
    # model.bfloat16()
    return model


def load_tokenizer(tokenizer_path):
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_path)
    return tokenizer

model_path = path
# model_path = "EleutherAI/gpt-neo-1.3B"
model = load_model(model_path)
tokenizer = load_tokenizer(model_path)
print("xd")

def generate_text(sequence, max_length):
    model_inputs = tokenizer([sequence], return_tensors="pt").to('cuda')
    final_outputs = model.generate(
        # ids.input_ids,
        **model_inputs,
        do_sample=True,
        max_length=max_length,
        pad_token_id=model.config.eos_token_id,
        top_k=50,
        top_p=0.95,
    )
    output = tokenizer.decode(final_outputs[0], skip_special_tokens=True)
    print(output)
    return output
```

```python
data_1 = ['Title: The Dalaran Fountain', 'Title: Ten Commendation Signets', 'Title: The Greater of Two Evils', 'Title: Hop Hunting', 'Title: Wayshrines of Zuldazar']
output = []
for title in data_1:
    output.append(generate_text(title, 800))

with open("test_model_gpt2_2.txt", "w", newline="\n") as file:
    for text in output:
      file.write(text)

print("done")
```
