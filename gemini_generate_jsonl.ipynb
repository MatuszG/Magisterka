```python
import csv
import json
import pandas as pd
import random
import os
from sklearn.model_selection import train_test_split

def csv_to_jsonl_with_split(csv_filename, train_jsonl_filename, val_jsonl_filename, test_size=0.2):
    # Load the CSV file into a pandas DataFrame
    csv_file = pd.read_csv(csv_filename)

    # Drop rows where 'Quest Description' is NaN
    csv_file = csv_file.dropna(subset=['Quest Description'])
    csv_file = csv_file.dropna(subset=['Quest Hand-in'])
    
    # Convert DataFrame back to list of dictionaries
    data = csv_file.to_dict(orient='records')
    
    
    # Split the data into training and validation sets
    train_data, val_data = train_test_split(data, test_size=test_size, random_state=42)

    # Function to write data to JSONL
    def write_jsonl(data, jsonl_filename):
        with open(jsonl_filename, mode='w') as jsonl_file:
            for row in data:
                message_content = f"Title: {row['Quest Title']}"
                json_obj = {
                    "messages": [
                        {
                            "role": "user",
                            "content": message_content
                        },
                        {
                            "role": "model",
                            "content": f"Description: {row['Quest Description']}\n\n Quest Hand-in: {row['Quest Hand-in']}"
                        },
                        # {
                        #     "role": "model",
                        #     "content": f"Description: {row['Quest Description']}"
                        # }
                    ]
                }
                jsonl_file.write(json.dumps(json_obj) + '\n')

    # Write the training data to the training JSONL file
    write_jsonl(train_data, train_jsonl_filename)

    # Write the validation data to the validation JSONL file
    write_jsonl(val_data, val_jsonl_filename)

# Usage example
csv_filename = 'data.csv'            # Replace with your CSV file name
train_jsonl_filename = os.path.join("Gemini Datasets", "gemini_train_data_wow_2.jsonl")       # Replace with your desired training JSONL file name
val_jsonl_filename = os.path.join("Gemini Datasets", "gemini_val_data_wow_2.jsonl")           # Replace with your desired validation JSONL file name

csv_to_jsonl_with_split(csv_filename, train_jsonl_filename, val_jsonl_filename, test_size=0.2)
```
